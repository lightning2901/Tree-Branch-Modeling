# aplicación de matrices de rotacion (ejemplo sintético)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
# Crear una cuadrícula de puntos
x = np.linspace(-1, 1, 5)
y = np.linspace(-1, 1, 5)
z = np.linspace(-1, 1, 5)
x, y, z = np.meshgrid(x, y, z)

# Convertir la cuadrícula a una lista de puntos
points = np.vstack([x.ravel(), y.ravel(), z.ravel()]).T


```

```python
#funciones (que hizo Leo)

def rotate_x(theta):
    return np.array([
        [1, 0, 0],
        [0, np.cos(theta), -np.sin(theta)],
        [0, np.sin(theta), np.cos(theta)]
    ])

def rotate_y(theta):
    return np.array([
        [np.cos(theta), 0, np.sin(theta)],
        [0, 1, 0],
        [-np.sin(theta), 0, np.cos(theta)]
    ])

def rotate_z(theta):
    return np.array([
        [np.cos(theta), -np.sin(theta), 0],
        [np.sin(theta), np.cos(theta), 0],
        [0, 0, 1]
    ])
```

```python
# ángulos de rotación (en radianes)
theta_x = np.pi / 4  # 45 grados
theta_y = np.pi / 6  # 30 grados
theta_z = np.pi / 3  # 60 grados

# matrices de rotación
R_x = rotate_x(theta_x)
R_y = rotate_y(theta_y)
R_z = rotate_z(theta_z)

# aplicar las rotaciones
rotated_points = points @ R_x.T 
rotated_points = rotated_points @ R_y.T  
rotated_points = rotated_points @ R_z.T 


```

```python
# Figura 1: Puntos originales y rotados
fig1 = go.Figure()

fig1.add_trace(go.Scatter3d(
    x=points[:, 0], y=points[:, 1], z=points[:, 2],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='Originales'
))

fig1.add_trace(go.Scatter3d(
    x=rotated_points[:, 0], y=rotated_points[:, 1], z=rotated_points[:, 2],
    mode='markers',
    marker=dict(size=5, color='red'),
    name='Rotados'
))

fig1.update_layout(
    title="Puntos Originales y Rotados",
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    ),
    legend=dict(x=0.8, y=0.9)
)

# Figura 2: Solo puntos originales
fig2 = go.Figure()

fig2.add_trace(go.Scatter3d(
    x=points[:, 0], y=points[:, 1], z=points[:, 2],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='Originales'
))

fig2.update_layout(
    title="Puntos Originales",
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    )
)

# Figura 3: Solo puntos rotados
fig3 = go.Figure()

fig3.add_trace(go.Scatter3d(
    x=rotated_points[:, 0], y=rotated_points[:, 1], z=rotated_points[:, 2],
    mode='markers',
    marker=dict(size=5, color='red'),
    name='Rotados'
))

fig3.update_layout(
    title="Puntos Rotados",
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    )
)


fig1.show()
```
